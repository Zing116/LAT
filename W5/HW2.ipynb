```python
import pandas as pd
import plotly.express as px

# 讀取CSV檔案
df = pd.read_csv('各校退學人數.csv')

# 繪製散佈圖
fig = px.scatter(df, x="在學學生數", y="退學人數小計", 
                 size="退學人數小計", color="設立別",
                 hover_name="學校名稱", log_x=True, size_max=60)

# 設置圖表標題和軸標籤
fig.update_layout(
    title="在學學生數和退學人數的關係",
    xaxis_title="在學學生數",
    yaxis_title="退學人數小計"
)

# 顯示圖表
fig.show()
```
