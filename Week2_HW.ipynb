```python
import pandas as pd

df = pd.read_csv('2017藝文經費補助.csv')

df.head()
```

```python
temp = df[['ComplexName','FValue']]
temp
```

```python
Type = temp.groupby(['ComplexName'])['FValue'].sum().reset_index()
```

```python
Type['year'] = 2017
Type
```

```python
df2016 = pd.read_csv('2018藝文經費補助.csv')
```

```python
new = df2016[['ComplexName','FValue']].groupby(['ComplexName'])['FValue'].sum().reset_index()
```

```python
new['year'] = 2018
```

```python
new
```

```python
frames = [Type, new]
result = pd.concat(frames)
```
